```python
import json
import matplotlib.pyplot as plt
from PIL import Image

# Path to the BDD100k dataset labels file
labels_path = './datasets/labels/train_A.json'

# Path to the image file you want to visualize
image_path = './datasets/bdd100k/train/A/70d7553c-7da414da.jpg'

# Load the labels file
with open(labels_path, 'r') as f:
    labels = json.load(f)

# Find the label for the image you want to visualize
image_name = image_path.split('/')[-1]
for item in labels:
    if item['name'] == image_name:
        label_data = item
        break

# Load the image file
image = Image.open(image_path)
#image = image.resize((512, 512))
row_scale = 512/1280
col_scale = 512/720

# Plot the image
plt.imshow(image)

# Plot the labels on top of the image
for obj in label_data['labels']:
    category = obj['category']
    bbox = obj['box2d']
    x1, y1, x2, y2 = bbox['x1']*row_scale, bbox['y1']*col_scale, bbox['x2']*row_scale, bbox['y2']*col_scale
    plt.plot([x1, x2], [y1, y1], color='red')
    plt.plot([x2, x2], [y1, y2], color='red')
    plt.plot([x2, x1], [y2, y2], color='red')
    plt.plot([x1, x1], [y2, y1], color='red')
    plt.text(x1, y1-10, category, color='red')

# Show the plot
plt.show()
```

```python
import os
from PIL import Image

import json

# read the JSON file into a dictionary
with open('./bdd100k_label/labels/det_20/det_val.json', 'r') as f:
    data = json.load(f)

# filter the dictionary to select only images with "timeofday" = "daytime"
daytime_images = [img for img in data if img['attributes']['timeofday'] == 'daytime']
night_images = [img for img in data if img['attributes']['timeofday'] == 'night']

# create a list of the names of the filtered images
daytime = [img['name'] for img in daytime_images]
night = [img['name'] for img in night_images]

# set the input and output folder paths
input_folder = "./bdd100k/images/100k/val/"
output_folder_day = "./bdd100k/images/100k/val_resized/A/"
output_folder_night = "./bdd100k/images/100k/val_resized/B/"

# loop through all the image files in the input folder
for filename in os.listdir(input_folder):
    if (filename in daytime) :
        # open the image file using PIL
        img_path = os.path.join(input_folder, filename)
        img = Image.open(img_path)

        # resize the image to 512x512
        img = img.resize((512, 512))

        # save the resized image to the output folder
        output_path = os.path.join(output_folder_day, filename)
        img.save(output_path)
        
    if (filename in night) : 
        # open the image file using PIL
        img_path = os.path.join(input_folder, filename)
        img = Image.open(img_path)

        # resize the image to 512x512
        img = img.resize((512, 512))

        # save the resized image to the output folder
        output_path = os.path.join(output_folder_night, filename)
        img.save(output_path)
```

```python
import os
import shutil
import random

source_folder = "./output_bdd100k/B"
list_folder = "./output_bdd100k/B"
source_txt_folder = "./datasets/bdd100k/train/A"
destination_folder = "./datasets/dataset_4/train/images"
destination_txt_folder = "./datasets/dataset_4/train/labels"
num_images_to_copy = 300

# Get a list of all the image files in the source folder
image_files = [filename for filename in os.listdir(list_folder) if filename.endswith('.jpg')]

# Randomly select num_images_to_copy image files
selected_image_files = random.sample(image_files, num_images_to_copy)

# Copy selected image files and their corresponding text files with the same filename to the destination folder
for image_file in selected_image_files:
    # Get the corresponding text file with the same filename as the image file
    text_file = os.path.splitext(image_file)[0] + ".txt"
    if os.path.isfile(os.path.join(source_txt_folder, text_file)):
        # Copy the image file and text file to the destination folder
        shutil.move(os.path.join(source_folder, image_file), os.path.join(destination_folder, image_file))
        shutil.copy(os.path.join(source_txt_folder, text_file), os.path.join(destination_txt_folder, text_file))
```

```python
from PIL import Image, ImageDraw, ImageFont

# Load the image
img = Image.open('./datasets/test_dataset/images/0a006b7b-ff7dae0c.jpg')

# Parse the annotation file
with open('./datasets/test_dataset/labels/0a006b7b-ff7dae0c.txt', 'r') as f:
    lines = f.readlines()

for line in lines:
    line = line.strip().split()
    class_id = int(line[0])
    x, y, w, h = map(float, line[1:])
    left = int((x - w/2) * img.width)
    top = int((y - h/2) * img.height)
    right = int((x + w/2) * img.width) 
    bottom = int((y + h/2) * img.height)

    # Draw the bounding box
    draw = ImageDraw.Draw(img)
    draw.rectangle(((left, top), (right, bottom)), outline=(0, 255, 0), width=2)
    #font = ImageFont.truetype('arial.ttf', 14)
    draw.text((left, top-14), str(class_id), fill=(0, 0, 255))

# Display the image
img.show()
```

```python
import os

categorys = ['bicycle', 'bus', 'car', 'motorcycle', 'pedestrian', 'rider',  'traffic light', 'traffic sign', 'train', 'truck']
class_label = ''

def yolo_to_voc(image_width, image_height, bbox):
    x, y, w, h = bbox
    dw = 1.0 / image_width
    dh = 1.0 / image_height

    # YOLO 좌표를 VOC 형식으로 변환
    voc_x = (x * image_width) - (w * image_width / 2)
    voc_y = (y * image_height) - (h * image_height / 2)
    voc_w = w * image_width
    voc_h = h * image_height

    # VOC 형식 좌표 반환 (xmin, ymin, xmax, ymax)
    xmin = max(0, voc_x)
    ymin = max(0, voc_y)
    xmax = min(image_width, voc_x + voc_w)
    ymax = min(image_height, voc_y + voc_h)

    return int(xmin), int(ymin), int(xmax), int(ymax)

def convert_labels_to_voc_format(source_folder, destination_folder, image_width, image_height):
    #os.makedirs(destination_folder, exist_ok=True)

    txt_files = [f for f in os.listdir(source_folder) if f.endswith(".txt")]

    for txt_file in txt_files:
        txt_path = os.path.join(source_folder, txt_file)
        with open(txt_path, "r") as file:
            lines = file.readlines()
        
        voc_lines = []
        for line in lines:
            #class_id, conf, x, y, w, h = map(float, line.strip().split())
            class_id, conf, x, y, w, h = map(float, line.strip().split())
            yolo_bbox = [x, y, w, h]
            voc_bbox = yolo_to_voc(image_width, image_height, yolo_bbox)
#             for i in range(len(categorys)) :
#                 if(int(class_id) == i) : 
#                     class_label = categorys[i]
#             voc_line = f"{class_label} {conf} {voc_bbox[0]} {voc_bbox[1]} {voc_bbox[2]} {voc_bbox[3]}\n"
            voc_line = f"{int(class_id)} {conf} {voc_bbox[0]} {voc_bbox[1]} {voc_bbox[2]} {voc_bbox[3]}\n"
            #voc_line = f"{int(class_id)} {voc_bbox[0]} {voc_bbox[1]} {voc_bbox[2]} {voc_bbox[3]}\n"

            voc_lines.append(voc_line)

        destination_path = os.path.join(destination_folder, txt_file)
        with open(destination_path, "w") as file:
            file.writelines(voc_lines)
        
        print(f"Converted {txt_file} to VOC format")

source_folder = "./datasets/test_dataset/input/labels"
destination_folder = "./datasets/test_dataset/input/detection-results"
image_width = 512  
image_height = 512 

convert_labels_to_voc_format(source_folder, destination_folder, image_width, image_height)
```
